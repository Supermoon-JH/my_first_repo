## 제어문

**비교문**

**if문**

단일 선택구조  
if ( 조건 )  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장1;  
if ( 조건 ) {  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장1;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장2;  
}  

이중 선택구조  
if ( 조건 )  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장1;  
else  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장2;  

다중 선택구조  
if ( 조건 )  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장1;  
else if ( 조건2 )  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장2;  
else  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장3;  

**switch ~ case문**

switch ( 변수 또는 수식 ){  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;case 값1:    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장1; break;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;case 값2:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장2; break;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;case 값3:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장3; break;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;case 값4:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장4; break;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;default:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문장5;  
}

**반복문**

**for문**

for ( 수식1(초기화) ; 수식2(조건) ; 수식3(인덱스증감) ){  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문1;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문2;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문N;  
}  

**while문**

while( 조건식 ){  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문 1;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문 2;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문 N;  
}  

**do~while문**(처음에 한번 수행하고 나서 조건을 비교하므로 최소한 한번은 수행)  

do{  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문 1;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문 2;  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행문 N;  
} while( 조건식 )  
 
**기타 제어문**
1)	break문: 선택구조나 반복구조를 더 이상 실행하지 말고 중단하여, 다음 명령문을 실행  
2)	continue문: 계속의 의미, 다음 반복구조로 다시 들어감
