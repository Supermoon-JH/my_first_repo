**서식을 가지는 입출력 함수**   
printf(“출력 서식”, 출력 대상);  
문자형: %c  
정수형: %d, l: long, u: unsigned, o: octo, x: hexa  
실수형: %f  
문자열형: %s  
  
**printf() 함수의 출력 서식에서 주의해야할 문자들**  
1.	기호”와 ‘는 출력서식의 “ “과 혼동할 수 있으므로, \”나 \’로 표현  
2.	% 기호는 %%로 표현  
3.	\는 \\로 표현  

**문자형 서식**  
\#include<stdio.h>  
void main(){  
printf(“12345\n”);  
printf(“%c\n”, ‘Y’);  
printf(“%5c\n”, ‘Y’);  
printf(“%-5c\n”, ‘Y’);  
}  

12345    
Y<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Y       
Y      

**정수형 서식**  
\#include<stdio.h>  
void main(){  
int x = 13579;  
printf(“12345\n”);  
printf(“%5d\n”, x);         # 5칸을 잡아서 오른쪽이 기준  
printf(“%3d\n”, x);  
printf(“%-7d\n”, x);        # 7칸을 잡아서 왼쪽이 기준  
printf(“%07d\n”, x);        # 7칸을 잡아서 오른쪽이 기준, 빈칸은 0으로 채움  
}    

12345  
13579  
13579  
13579  
0013579  

**실수형 서식**  
\#include<stdio.h>  
void main(){  
float x = 13.579;  
printf(“12345678\n”);  
printf(“%f\n”, x);        # 오른쪽이 기준, 소수점 6자리가 default, .도 자리 차지  
printf(“%8.2f\n”, x);     # 8칸을 잡고 오른쪽이 기준, 소수점 2자리  
printf(“%-8.2f\n”, x);    # 8칸을 잡고 왼쪽이 기준, 소수점 2자리  
printf(“%08.2f\n”, x);    # 8칸을 잡고 오른쪽이 기준, 소수점 2자리, 빈칸은 0  
}

12345678  
13.579000<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;13.58   
13.58  
00013.58
